```python
df['Release_Date']=pd.to_datetime(df['Release_Date'])
print(df['Release_Date'].dtypes)
```

```python
df['Release_Date'] = df['Release_Date'].dt.year
df['Release_Date'].dtypes
```

```python
df.info()
```

```python
# making list of column to be dropped
cols = ['Overview', 'Original_Language', 'Poster_Url']
# dropping columns and confirming changes
df.drop(cols, axis = 1, inplace = True)
df.columns
```

```python
df.head()
```

```python
#categorizing Vote_Average column
def catigorize_col (df, col, labels):
    edges = [df[col].describe()['min'],
             df[col].describe()['25%'],
             df[col].describe()['50%'],
             df[col].describe()['75%'],
             df[col].describe()['max']]
    df[col] = pd.cut(df[col], edges, labels = labels, duplicates='drop')
    return df
```

```python
# define labels for edges
labels = ['not_popular', 'below_avg', 'average', 'popular']
# categorize column based on labels and edges
catigorize_col(df, 'Vote_Average', labels)
# confirming changes
df['Vote_Average'].unique()
```

```python
df.head()
```

```python
# exploring column
df['Vote_Average'].value_counts()
```

```python
# dropping NaNs
df.dropna(inplace = True)
# confirming
df.isna().sum()
```

```python
# split the strings into lists
df['Genre'] = df['Genre'].str.split(', ')
# explode the lists
df = df.explode('Genre').reset_index(drop=True)
df.head()
```

```python
# casting column into category
df['Genre'] = df['Genre'].astype('category')
# confirming changes
df['Genre'].dtypes
```

```python
df.info()
```

```python
df.nunique()
```

```python
# setting up seaborn configurations
sns.set_style('whitegrid')
```

```python
# What is the most frequent genre in the data set?
```

```python
# showing stats. on genre column
df['Genre'].describe()
```

```python
# visualizing genre column
sns.catplot(y = 'Genre', data = df, kind = 'count',
 order = df['Genre'].value_counts().index,
 color = '#4287f5')
plt.title('genre column distribution')
plt.show()
```

```python
# we can notice from the above visual that Drama genre is the most frequent genrein our dataset and has appeared more than 14% of the times among 19 othergenres.
```

```python
# What genres has highest votes?
sns.catplot(y = 'Vote_Average', data = df, kind = 'count',
 order = df['Vote_Average'].value_counts().index,
 color = '#4287f5')
plt.title('votes destribution')
plt.show()
```

```python
# Which movie got the highest popularity? What is it's genre?
df[df['Popularity'] == df['Popularity'].max()]
```

```python
# Which movie got the lowest popularity? What is it's genre?
df[df['Popularity'] == df['Popularity'].min()]
```

```python
# Which year has most filmmed movies?
df['Release_Date'].hist()
plt.title('Release_Date column distribution')
plt.show()
```

```python
# Which genre has highest average vote count?
genre_votes = df.groupby('Genre')['Vote_Count'].mean().sort_values(ascending=False)

genre_votes.head(10)
genre_votes.head(10).plot(kind='bar')
plt.title("Top Genres by Average Vote Count")
plt.show()
```

```python
# Trend of movies over time(year wise growth)
movies_per_year = df.groupby('Release_Date')['Title'].nunique()
movies_per_year.plot()
plt.title("Movies Released Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Movies")
plt.show()
```

```python
# correlation between popularity and vote count
df[['Popularity', 'Vote_Count']].corr()
sns.heatmap(df[['Popularity', 'Vote_Count']].corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
# distribution of vote count
sns.histplot(df['Vote_Count'], bins=50)
plt.title("Vote Count Distribution")
plt.show()
```

```python
# top 5 genre per year
top_genre_per_year = df.groupby(['Release_Date', 'Genre'])['Popularity'].mean().reset_index()

top_genre_per_year.sort_values(['Release_Date', 'Popularity'], ascending=[True, False]).head(20)
```

```python
# which genre contributes most to total popularity
genre_popularity_sum = df.groupby('Genre')['Popularity'].sum().sort_values(ascending=False).head(6)

plt.figure()
plt.pie(genre_popularity_sum,
        labels=genre_popularity_sum.index,
        autopct='%1.1f%%')
plt.title("Popularity Contribution by Genre")
plt.show()
```

```python
# market share by year
recent_years = df[df['Release_Date'] >= df['Release_Date'].max()-4]

year_share = recent_years['Release_Date'].value_counts()

plt.figure()
plt.pie(year_share,
        labels=year_share.index,
        autopct='%1.1f%%')
plt.title("Movie Share in Last 5 Years")
plt.show()
```
